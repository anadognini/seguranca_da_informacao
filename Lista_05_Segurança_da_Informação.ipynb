```python
# Questão 1

# Nesta questão iremos exercitar a implementação do algoritmo de cifragem
# denominado de “Cifra de transposição geométrica colunar”. Para isso,
# construa um programa que possibilite ao usuário cifrar algum texto e decifrar
# algum texto utilizando este algoritmo. Lembre-se que para realizar a cifragem
#  e a decifragem, o usuário deve poder informar como parâmetro a quantidade de
# colunas da matriz. Considere que o texto simples irá conter apenas letras e
# espaços. O algoritmo de cifragem deve suprimir os caracteres em branco.

def cifrar(texto, colunas):
    texto = texto.replace(" ", "")
    num_linhas = -(-len(texto) // colunas)

    matriz = ['' for _ in range(num_linhas)]
    idx = 0
    for i in range(num_linhas):
        for j in range(colunas):
            if idx < len(texto):
                matriz[i] += texto[idx]
                idx += 1

    texto_cifrado = ''
    for j in range(colunas):
        for i in range(num_linhas):
            if j < len(matriz[i]):
                texto_cifrado += matriz[i][j]

    return texto_cifrado

def decifrar(texto_cifrado, colunas):
    num_linhas = -(-len(texto_cifrado) // colunas)

    matriz = ['' for _ in range(num_linhas)]
    idx = 0
    for j in range(colunas):
        for i in range(num_linhas):
            if idx < len(texto_cifrado):
                matriz[i] += texto_cifrado[idx]
                idx += 1

    texto_decifrado = ''.join(matriz)

    return texto_decifrado

def main():
    escolha = input("Escolha a operação (cifrar/decifrar): ").strip().lower()
    texto = input("Digite o texto: ").strip()
    colunas = int(input("Digite o número de colunas: "))

    if escolha == "cifrar":
        texto_cifrado = cifrar(texto, colunas)
        print(f"Texto cifrado: {texto_cifrado}")
    elif escolha == "decifrar":
        texto_decifrado = decifrar(texto, colunas)
        print(f"Texto decifrado: {texto_decifrado}")
    else:
        print("Opção inválida! Escolha 'cifrar' ou 'decifrar'.")

if __name__ == "__main__":
    main()
```

```python
# Questão 2
# Implemente o algoritmo de cifragem “Cerca ferroviária”. O algoritmo deve
# permitir que o usuário submeta um texto a ser cifrado. O programa também
# deve permitir decifrar textos com este algoritmo. Este algoritmo requer que o
# usuário forneça como entrada a quantidade de trilhos a ser utilizada.
# Considere que o texto simples irá conter apenas letras e espaços. O algoritmo
# de cifragem deve suprimir os caracteres em branco.

def cifrar(texto, trilhos):
    texto = texto.replace(" ", "")
    matriz = [['' for _ in range(len(texto))] for _ in range(trilhos)]

    i, direção = 0, 1
    for char in texto:
        matriz[i].append(char)
        if i == 0: direção = 1
        elif i == trilhos - 1: direção = -1
        i += direção

    texto_cifrado = ''.join([''.join(linha) for linha in matriz])

    return texto_cifrado

def decifrar(texto_cifrado, trilhos):
    num_caracteres = len(texto_cifrado)
    matriz = [['' for _ in range(num_caracteres)] for _ in range(trilhos)]

    i, direção = 0, 1
    indice = 0
    for col in range(num_caracteres):
        matriz[i][col] = texto_cifrado[indice]
        indice += 1
        if i == 0: direção = 1
        elif i == trilhos - 1: direção = -1
        i += direção

    texto_decifrado = ''
    for i in range(trilhos):
        for j in range(num_caracteres // trilhos):
            if matriz[i][j] != '':
                texto_decifrado += matriz[i][j]

    return texto_decifrado

def main():
    escolha = input("Escolha a operação (cifrar/decifrar): ").strip().lower()
    texto = input("Digite o texto: ").strip()
    trilhos = int(input("Digite o número de trilhos: "))

    if escolha == "cifrar":
        texto_cifrado = cifrar(texto, trilhos)
        print(f"Texto cifrado: {texto_cifrado}")
    elif escolha == "decifrar":
        texto_decifrado = decifrar(texto, trilhos)
        print(f"Texto decifrado: {texto_decifrado}")
    else:
        print("Opção inválida! Escolha 'cifrar' ou 'decifrar'.")

if __name__ == "__main__":
    main()
```
